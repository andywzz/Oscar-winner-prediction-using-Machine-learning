````python
### Add note about outlier detection ###
#2018 busiest traffic station = Times SQ 42-st
#204,017 ppl per day
#50 turnstiles in this station
#4 hour intervals
#10x the average turnstile in the busiest station

(204017/50)/6*10
````

````python
def get_data(week_nums):
    """
    
    Forms a dataframe from specified MTA data weeks.
    
    """
    
    url = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_{}.txt"
    dfs = []
    for week_num in week_nums:
        file_url = url.format(week_num)
        dfs.append(pd.read_csv(file_url))
    return pd.concat(dfs)
        
week_nums = [190223, 190302, 190309, 190316, 190323, 190330, 190406, 190413, 190420, 190427, 190504, 190511, 190518,
            190525, 190601, 200229, 200307, 200314, 200321, 200328, 200404, 200411, 200418, 200425, 200502, 200509,
            200516, 200523, 200530, 200606, 200613, 200620, 200627]
turnstiles_df = get_data(week_nums)
````

````python
def column_clean(df):
    """
    
    Converts the 'date' and 'time' columns into its own
    datetime formatted column, as well as separates this
    into year, week, day of the week, month, and hour.
    
    """
    
    #Datetime column
    df['date_time'] = pd.to_datetime(df['date'] +' '+ df['time'])
    
    #Extracting parts of date/time
    df['year'] = df['date_time'].dt.year
    df['week'] = df['date_time'].dt.week
    df['day'] = df['date_time'].dt.dayofweek
    df['month'] = df['date_time'].dt.month
    df['hour'] = df['date_time'].dt.hour
    df['week_num'] = (df['date_time'].dt.day - 1) // 7 + 1
    
    return df[(((df['date_time'] >= '2019-03-01') & (df['date_time'] <= '2019-07-01')) | 
               (((df['date_time'] >= '2020-03-01') & (df['date_time'] <= '2020-07-01'))))]
````

````python
turnstile_col_clean = column_clean(turns_copy)
````

````python
turnstile_col_clean.head()
````

To handle anomalies in the data, new counts are then processed. First, 3 criteria are evaluated to determine if the raw entry count is unreliable:
1. Entries > 6801 (10x average weekday turnstile in busiest station)
    * Based off of the 2018 MTA average weekday ridership data (http://web.mta.info/nyct/facts/ridership/ridership_sub.htm)
        * Highest daily ridership station = Times sq 42 st
        * 50 unique turnstiles in this station
        * 6 time intervals per day
        * 6801 = ```((weekday daily average Times sq 42 st)/(50 turnstiles)/(6 intervals)) * 10```
2. Turnstiles less than 10200.85 but greater than 100x the previous recorded entry 
3. Turnstiles <= 0 entries

If an outlier is found, is then replaced with that turnstile's weekly median entry count - if not, it is left alone.

````python
#Raw exit counts of differences between subsequent counts
turnstile_col_clean['exit_counts'] = (
    turnstile_col_clean.groupby(['station', 'unit', 'scp', 'date'], as_index = False)[['exits']].transform(pd.Series.diff) 
                                )
````

````python
#Processed exit counts
turnstile_col_clean['new_exits'] = (
    np.where(((turnstile_col_clean['exit_counts'] > 6801) | 
              (turnstile_col_clean['exit_counts'] > 100*turnstile_col_clean['exit_counts'].shift()) |
              (turnstile_col_clean['exit_counts'] <= 0)),
             (np.asarray(turnstile_col_clean.groupby(['station', 'unit', 'scp', 'year', 'month'], 
                                                as_index = False)[['exit_counts']]
                         .transform(pd.Series.median))
              .reshape(len(turnstile_col_clean),)
             ),
             turnstile_col_clean['exit_counts'])
                              )
````

````python
#Raw entrance counts of differences between subsequent counts
turnstile_col_clean['entry_counts'] = (
    turnstile_col_clean.groupby(['station', 'unit', 'scp', 'date'], as_index = False)[['entries']].transform(pd.Series.diff) 
                                )
````

````python
#Processed entry counts
turnstile_col_clean['new_entry'] = (
    np.where(((turnstile_col_clean['entry_counts'] > 6801) | 
              (turnstile_col_clean['entry_counts'] > 100*turnstile_col_clean['entry_counts'].shift()) |
              (turnstile_col_clean['entry_counts'] <= 0)),
             (np.asarray(turnstile_col_clean.groupby(['station', 'unit', 'scp', 'year', 'month'], 
                                                as_index = False)[['entry_counts']]
                         .transform(pd.Series.median))
              .reshape(len(turnstile_col_clean),)
             ),
             turnstile_col_clean['entry_counts'])
                              )
````

# Merging in Manhattan data

````python
manhattan_stations = pd.read_csv('manhattan_station_nta.csv')
````

````python
manhattan_stations.head()
````

````python
#Inner merge to filter out non-manhattan stations
manhattan_turnstiles = turnstile_col_clean.merge(manhattan_stations, left_on = 'station', right_on = 'Station')
````

````python
manhattan_daily_stations = (manhattan_turnstiles.groupby(['station', 'date'])[['new_entry', 'new_exits']]
                            .sum()
                            .reset_index()
                            .merge(manhattan_stations, left_on = 'station', right_on = 'Station')
                            .drop('Station', axis = 1))
````

````python
manhattan_daily_stations.head()
````

````python
#Remerging in parsed date specs
manhattan_daily_clean = (manhattan_daily_stations.merge(turnstile_col_clean[['station', 'date', 'year', 
                                                                             'month', 'week_num', 'day']]
                                                        .drop_duplicates())
                        [['GeoID', 'GeogName', 'station', 'date', 'year', 
                          'month', 'week_num', 'day', 'new_entry', 'new_exits']])
manhattan_daily_clean.head()
````

# Manhattan 2019-2020 comparisons

````python
manhattan_daily_clean['date_time'] = pd.to_datetime(manhattan_daily_clean['date'])

manhattan_correct_time = (
    
    manhattan_daily_clean[(((manhattan_daily_clean['date_time'] >= '2019-03-01') & 
                        (manhattan_daily_clean['date_time'] < '2019-05-01')) | 
                       (((manhattan_daily_clean['date_time'] >= '2020-03-01') & 
                         (manhattan_daily_clean['date_time'] < '2020-05-01'))))]
)
````

## Average daily ridership per week

````python
neighborhood_week = (manhattan_correct_time.groupby(['GeogName', 'year', 'month', 'week_num'], 
                                                    as_index = False)[['new_entry', 'new_exits']].mean())
neighborhood_week.head()
````

````python
neighborhood_week_compare = neighborhood_week[neighborhood_week['year'] == 2019].merge(neighborhood_week[neighborhood_week['year'] == 2020],
                                                                                   left_on = ['GeogName', 'month', 'week_num'],
                                                                                   right_on = ['GeogName', 'month', 'week_num'],
                                                                                   suffixes = ('_19', '_20'))
````

````python
neighborhood_week_compare['entry_pct_change'] = (neighborhood_week_compare[['new_entry_19', 'new_entry_20']]
                                                 .pct_change(axis = 1)['new_entry_20']).apply(lambda x: x*100)

neighborhood_week_compare['exit_pct_change'] = (neighborhood_week_compare[['new_exits_19', 'new_exits_20']]
                                                 .pct_change(axis = 1)['new_exits_20']).apply(lambda x: x*100)
````

````python
neighborhood_week_compare.head()
````

````python
mean_week_changes = (neighborhood_week_compare.groupby(['GeogName'], as_index = False)
                     [['entry_pct_change', 'exit_pct_change']]
                     .mean()
                     .sort_values('entry_pct_change'))

#mean_week_changes.head(10).to_csv('top_10_nta_changes.csv')
````

````python
mean_week_changes.head(10)
````

````python
top_5_weekly = (neighborhood_week_compare.groupby(['GeogName'], as_index = False)[['entry_pct_change', 'exit_pct_change']]
                .mean()
                .sort_values('entry_pct_change')
                .head()['GeogName'].tolist())
````

## Average weekly ridership per month

````python
#Average weekly ridership per month
neighborhood_month = (manhattan_correct_time.groupby(['GeogName', 'year', 'month', 'week_num'], as_index = False)[['new_entry', 'new_exits']]
                      .sum()
                      .groupby(['GeogName', 'year', 'month'], as_index = False)[['new_entry', 'new_exits']]
                      .mean())
neighborhood_month.head()
````

````python
neighborhood_month_compare = neighborhood_month[neighborhood_month['year'] == 2019].merge(neighborhood_month[neighborhood_month['year'] == 2020],
                                                                                   left_on = ['GeogName', 'month'],
                                                                                   right_on = ['GeogName', 'month'],
                                                                                   suffixes = ('_19', '_20'))
````

````python
neighborhood_month_compare['entry_pct_change'] = (neighborhood_month_compare[['new_entry_19', 'new_entry_20']]
                                                 .pct_change(axis = 1)['new_entry_20'])

neighborhood_month_compare['exit_pct_change'] = (neighborhood_month_compare[['new_exits_19', 'new_exits_20']]
                                                 .pct_change(axis = 1)['new_exits_20'])
````

````python
neighborhood_month_compare.head()
````

````python
(neighborhood_month_compare.groupby(['GeogName'], as_index = False)[['entry_pct_change', 'exit_pct_change']].mean().sort_values('entry_pct_change'))
````

````python
top_5_monthly = (neighborhood_month_compare.groupby(['GeogName'], as_index = False)[['entry_pct_change', 'exit_pct_change']]
                .mean()
                .sort_values('entry_pct_change')
                .head()['GeogName'].tolist())
````

````python
#Weekly date ranges
date_ranges = [f'3/{i}' for i in range(1,31,7)] + [f'4/{i}' for i in range(1,30,7)]
date_ranges
````

## Raw ridership differences

### Subplots

````python
plt.figure(figsize=[16,8])

for idx,neighborhood in enumerate(top_5_weekly):
    plt.subplot(2,3,idx+1)
    plt.plot(date_ranges, (np.array(neighborhood_week[(neighborhood_week['GeogName'] == neighborhood) &
                                              (neighborhood_week['year'] == 2020)]['new_entry']) - 
                          np.array(neighborhood_week[(neighborhood_week['GeogName'] == neighborhood) &
                                              (neighborhood_week['year'] == 2019)]['new_entry'])))

    plt.title(neighborhood)
    plt.ylabel('Ridership difference')
    plt.xlabel('Week')

plt.suptitle("Average Daily Ridership Differences", weight='bold', fontsize=14)
plt.subplots_adjust(left = .04, hspace=.5)
````

````python
plt.figure(figsize=[16,8])

for idx,neighborhood in enumerate(top_5_weekly):
    #plt.subplot(2,3,idx+1)
    
    baseline_mean = np.array(neighborhood_week[(neighborhood_week['GeogName'] == neighborhood) &
                                              (neighborhood_week['year'] == 2019)]['new_entry']).mean()
    
    plt.plot(date_ranges, (np.array(neighborhood_week[(neighborhood_week['GeogName'] == neighborhood) &
                                              (neighborhood_week['year'] == 2020)]['new_entry'])/baseline_mean 
                          ), 
            label = neighborhood)

plt.legend()
plt.title('Average Daily Ridership Differences (DIVIDED BY SOME_VALUE)')
plt.ylabel('Ridership difference')
plt.xlabel('Week')
````

````python
plt.figure(figsize=[16,8])
plt.style.use('seaborn')
for idx,neighborhood in enumerate(top_5_weekly):
    #plt.subplot(2,3,idx+1)
    
    baseline_mean = np.array(neighborhood_week[(neighborhood_week['GeogName'] == neighborhood) &
                                              (neighborhood_week['year'] == 2019)]['new_entry']).mean()
    baseline_sd = np.array(neighborhood_week[(neighborhood_week['GeogName'] == neighborhood) &
                                              (neighborhood_week['year'] == 2019)]['new_entry']).std()
    
    plt.plot(date_ranges, (np.array(neighborhood_week[(neighborhood_week['GeogName'] == neighborhood) &
                                              (neighborhood_week['year'] == 2020)]['new_entry']) - baseline_mean)/baseline_sd
                          , 
            label = neighborhood)
    
plt.axvline(1, 0, 2, color = 'r', linestyle = '--', alpha = .5, label = 'state of emergency')
plt.legend()
plt.title('Average Daily Ridership Differences', fontsize = 16, weight = 'bold')
plt.ylabel('Ridership difference (z-scored)', fontsize = 14)
plt.xlabel('Week', fontsize = 14)
plt.savefig('zscore_daily_diffs_sns_v2.png');
````
