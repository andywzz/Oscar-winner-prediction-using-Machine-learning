This notebook contains code used to create the plots found in our presentation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_rows',400)
from IPython.core.display import display, HTML
#display(HTML("<style>.container { width:100% !important; }</style>"))

import os
from matplotlib import font_manager as fm
import matplotlib.patches as mpatches
from matplotlib.lines import Line2D


prop = fm.FontProperties(fname = #'yourpath/Fonts/JosefinSans-Regular.ttf', size=20)
legend_prop = fm.FontProperties(fname = #'yourpath/Fonts/JosefinSans-Regular.ttf', size=30)
legend_prop_small = fm.FontProperties(fname = #'yourpath/Fonts/JosefinSans-Regular.ttf', size=30)
```

```python
"""
Reading in datasets
"""

manhattan_demographics = pd.read_csv("manhattan_demographics_data.csv")
manhattan_housing_densities = pd.read_csv("manhattan_housing_densities.csv")
top_10_housing_densities = pd.read_csv("top_10_manhattan_housing_densities.csv")
top_10_demographics = pd.read_csv("top_10_manhattan_demographics.csv")
top_5_demo = pd.read_csv("top_5_manhattan_demographics.csv")
top_5_housing = pd.read_csv("top_5_manhattan_housing_densities.csv")
```

```python
"""
Re-index data in order of increasing average decrease
"""


top_5_housing = top_5_housing.reindex([2,3,4,0,1])
top_5_demo = top_5_demo.reindex([2,3,4,0,1])
```

```python
"""
Calculating percent population in target demographics for top 5
"""

top_5_demo['target_age'] = top_5_demo.age_18_29 + top_5_demo.age_30_49
top_5_demo['target_income'] = top_5_demo.income_30_74 + top_5_demo.income_75_plus
top_5_demo['target_edu'] = top_5_demo.edu_some_college + top_5_demo.edu_college_plus
```

```python
"""
Calculating percent population in target demographics for all 
"""

manhattan_demographics['target_age'] = manhattan_demographics.age_18_29 + manhattan_demographics.age_30_49
manhattan_demographics['target_income'] = manhattan_demographics.income_30_74 + manhattan_demographics.income_75_plus
manhattan_demographics['target_edu'] = manhattan_demographics.edu_some_college + manhattan_demographics.edu_college_plus
```

```python
"""
Get mean values for target demographics
"""

mean_target_age = manhattan_demographics.target_age.mean()
mean_target_edu = manhattan_demographics.target_edu.mean()
mean_target_income = manhattan_demographics.target_income.mean()
```

```python
"""
Demographics plot for all neighborhoods
"""

plt.figure(figsize=[10,10])
legend_prop_small = fm.FontProperties(fname = '/Users/kaitlin/Documents/Josefin_Sans/static/JosefinSans-Regular.ttf', size=30)


ind = np.arange(len(manhattan_demographics))[::-1]
label_arr = np.array(manhattan_demographics.GeogName)
y1 = np.array(manhattan_demographics.target_age)
y2 = np.array(manhattan_demographics.target_edu)
y3 = np.array(manhattan_demographics.target_income)

legend_arr = ['Target Age', 'Target Income', 'Target Education']
width = 0.2

plt.barh(ind, y1, width, color='#ae5e75')
plt.barh(ind + width, y3, width, color='#b4b4b4')
plt.barh(ind + 2 *width, y2, width, color='#ed7c43')

plt.xlim(0,100)
plt.ylim(-0.5, len(manhattan_demographics) - 0.5)
plt.xlabel("Percent Population",fontsize=15,fontproperties=prop, size=30, labelpad=20)
plt.margins(y=.1, x=.1)
plt.grid(axis='x',alpha=0.6,zorder=0)
plt.xticks(np.linspace(0,100,11), fontproperties=prop, size=25)
plt.yticks(ticks = ind, labels=label_arr, fontproperties=prop, size=25)
plt.box(on=None)


pink = mpatches.Patch(color='#ae5e75', label='Target Age')
grey = mpatches.Patch(color='#b4b4b4', label='Target Income')
orange = mpatches.Patch(color='#ed7c43', label='Target Education')

plt.legend(handles=[pink,grey,orange], prop=legend_prop_small, frameon=False,bbox_to_anchor=(.63,1.35), loc='upper right')
```

```python
"""
Demographics plot for top 5 neighborhoods
"""

plt.figure(figsize=[10,10])
legend_prop_small = fm.FontProperties(fname = '/Users/kaitlin/Documents/Josefin_Sans/static/JosefinSans-Regular.ttf', size=30)


ind = np.arange(len(top_5_demo))[::-1]
label_arr = np.array(top_5_demo.GeogName)
y1 = np.array(top_5_demo.target_age)
y2 = np.array(top_5_demo.target_edu)
y3 = np.array(top_5_demo.target_income)

legend_arr = ['Target Age', 'Target Income', 'Target Education']
width = 0.2

plt.barh(ind, y1, width, color='#ae5e75')
plt.barh(ind + width, y3, width, color='#b4b4b4')
plt.barh(ind + 2 *width, y2, width, color='#ed7c43')

plt.xlim(0,100)
plt.ylim(-0.5, len(top_5_demo) - 0.5)
plt.xlabel("Percent Population",fontsize=15,fontproperties=prop, size=30, labelpad=20)
plt.margins(y=.1, x=.1)
plt.grid(axis='x',alpha=0.6,zorder=0)
plt.xticks(np.linspace(0,100,11), fontproperties=prop, size=25)
plt.yticks(ticks = ind, labels=label_arr, fontproperties=prop, size=25)
plt.box(on=None)

plt.plot(mean_target_age, 4,'X',color='black', markersize = 13, mew=.25)
plt.plot(mean_target_age, 3,'X',color='black', markersize = 13, mew=.25)
plt.plot(mean_target_age, 2,'X',color='black', markersize = 13, mew=.25)
plt.plot(mean_target_age, 1,'X',color='black', markersize = 13, mew=.25)
plt.plot(mean_target_age, 0,'X',color='black', markersize = 13, mew=.25)

plt.plot(mean_target_income, 4.2,'X',color='black', markersize = 13, mew=.25)
plt.plot(mean_target_income, 3.2,'X',color='black', markersize = 13, mew=.25)
plt.plot(mean_target_income, 2.2,'X',color='black', markersize = 13, mew=.25)
plt.plot(mean_target_income, 1.2,'X',color='black', markersize = 13, mew=.25)
plt.plot(mean_target_income, 0.2,'X',color='black', markersize = 13, mew=.25)

plt.plot(mean_target_edu, 4.4,'X',color='black', markersize = 13, mew=.25)
plt.plot(mean_target_edu, 3.4,'X',color='black', markersize = 13, mew=.25)
plt.plot(mean_target_edu, 2.4,'X',color='black', markersize = 13, mew=.25)
plt.plot(mean_target_edu, 1.4,'X',color='black', markersize = 13, mew=.25)
plt.plot(mean_target_edu, 0.4,'X',color='black', markersize = 13, mew=.25)

pink = mpatches.Patch(color='#ae5e75', label='Target Age')
grey = mpatches.Patch(color='#b4b4b4', label='Target Income')
orange = mpatches.Patch(color='#ed7c43', label='Target Education')
x = Line2D([0], [0], marker = 'X', markersize=25, color='w', lw=4, label='Mean Value for Manhattan', markerfacecolor='black')

plt.legend(handles=[pink,grey,orange,x], prop=legend_prop_small, frameon=False,bbox_to_anchor=(.88,1.45), loc='upper right')
plt.show()
```

```python
"""
Plot of housing density
"""

plt.figure(figsize=[15,10])

r = np.array(top_5_housing.NTA)[::-1]
bars1 = np.array(top_5_housing.rel_occ_high_density)[::-1]
bars2 = np.array(top_5_housing.rel_occ_med_density)[::-1]
bars3 = np.array(top_5_housing.rel_occ_low_density)[::-1]
#grey = b4b4b4
#pink = c543a6
#puple = ae5e75
#orange = ed7c43
#dark purple = 6f40a8
colors = ['#ae5e75','#ed7c43','#b4b4b4']
labels = ['High', 'Medium', 'Low']
legend_prop_small = fm.FontProperties(fname = #'yourpath/Fonts/JosefinSans-Regular.ttf', size=20)

barWidth = 0.65
lefts = 0
for bars, col, label in zip([bars1, bars2, bars3], colors, labels):
    plt.barh(r, bars, left=lefts, color=col, edgecolor='white', height=barWidth, label=label,zorder=3)
    lefts += bars
plt.legend(fontsize=20)
plt.xlim(0,2000000)
plt.ylim(-0.5, len(bars) - 0.5)
plt.xlabel("Occupied Units per Square Mile",fontsize=15,fontproperties=prop, size=20, labelpad=25)
plt.margins(y=.1, x=.1)
plt.xticks(np.linspace(0,2000000,6), fontproperties=prop, size=20)
plt.yticks(np.array(top_5_housing.NTA), fontproperties=prop, size=20)
plt.box(on=None)
leg =plt.legend(prop=legend_prop_small, frameon=False,bbox_to_anchor=(0.49,1.4), loc='upper right', title="Housing Density", title_fontsize=25)
leg._legend_box.align='left'
```
